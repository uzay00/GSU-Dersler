# Gaussian Dağılımının Şifreleri

$$
p(x; \theta, \sigma^2) = \frac{1}{\sqrt{2\pi} \sigma} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-11, 11, 45)
```

```python
x
```

```python
plt.plot(x, np.exp(x))
```

```python
plt.plot(x, np.exp(-x))
```

```python
plt.plot(x, np.exp(-x**2))
```

```python
plt.plot(x, np.exp(-x**2/10))
```

```python
plt.plot(x, np.exp(-(x-5)**2/10))
```

```python
plt.plot(x, np.exp(-x**2), label = r'$e^{-x^2}$')
plt.plot(x, np.exp(-x**2/10), label = r'$e^{-\frac{x^2}{10}}$')
plt.plot(x, np.exp(-(x-5)**2/10), label = r'$e^{-\frac{(x-5)^2}{10}}$')
plt.legend()
plt.grid()
```
